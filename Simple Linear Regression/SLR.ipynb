# Step 3 - Seperate X(RND) and y(PROFIT)

```python
X = df[['RND']]
Y = df[['PROFIT']]
```

```python
X.head()
```

```python
Y.head()
```

# Step 4 - Build The Linear Regerssion Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

# PROFIT = B0 + B1*RND

# B0 - Intercept
# B1 - Slope/Coefficient

```python
model.intercept_
```

```python
model.coef_
```

# PROFIT_pred = 49032.9 + 0.8542*RND
# If RND increases by 1 unit Profit will increases by 0.8542
# if RND budget is 0 then the PROFIT is 49032.9 on an average

# Step 5 - Model evaluation


```python
ypred = model.predict(X)
ypred[0:5]
```

```python
Y.head()
```

```python
from sklearn.metrics import(
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
```

```python
def evaluate_model(model,x,y):
    # Predict the results on x
    ypred = model.predict(x)
    
    # Evaluate all metrics
    rmse = root_mean_squared_error(y,ypred)
    mae = mean_absolute_error(y,ypred)
    mape = mean_absolute_percentage_error(y,ypred)
    r2 = r2_score(y,ypred)

    # Print above results
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE :{mae:.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 score : {r2:.2%}")
```

```python
evaluate_model(model,X,Y)
```

# From above R2 score > 80% hence model is good
